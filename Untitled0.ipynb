```python
import pandas as pd

filename = 'train.csv'
df = pd.read_csv(filename, parse_dates=True, infer_datetime_format=True)
df.head(5)
```

```python
print("Rows,Cols:", df.shape)
print("\nColumns:\n", df.columns)
print("\nData types:\n", df.dtypes)
print("\nMissing values per column:\n", df.isna().sum())
```

```python
date_col = 'OrderDate'   # change this if your column has different name

df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
df = df.dropna(subset=[date_col])   # drop rows where date couldn't parse

# create Month and Year columns
df['Month'] = df[date_col].dt.to_period('M').astype(str)
df['Year'] = df[date_col].dt.year

# Show sample
df[[date_col, 'Month', 'Year']].head()
```

```python
date_col = 'Order Date'   # change this if your column has different name

# try to parse the date column
df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
df = df.dropna(subset=[date_col])   # drop rows where date couldn't parse

# create Month and Year columns
df['Month'] = df[date_col].dt.to_period('M').astype(str)
df['Year'] = df[date_col].dt.year

# Show sample
df[[date_col, 'Month', 'Year']].head()
```

```python
sales_col = 'Sales'
# try to convert to numeric
df[sales_col] = pd.to_numeric(df[sales_col], errors='coerce')
print("Sales missing after convert:", df[sales_col].isna().sum())
df = df.dropna(subset=[sales_col])  # drop rows where sales couldn't be read
df[sales_col] = df[sales_col].astype(float)
df[sales_col].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group monthly sales
monthly = df.groupby('Month')[sales_col].sum().reset_index()
plt.figure(figsize=(10,4))
sns.lineplot(data=monthly, x='Month', y=sales_col, marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Revenue')
plt.tight_layout()
plt.show()

# Save figure to file
plt.savefig('monthly_revenue.png', bbox_inches='tight')
print("Saved monthly_revenue.png")
```

```python
prod_col = 'Product Name'   # change if your dataset uses different name
top10 = df.groupby(prod_col)[sales_col].sum().reset_index().sort_values(sales_col, ascending=False).head(10)

plt.figure(figsize=(8,5))
sns.barplot(data=top10, y=prod_col, x=sales_col)
plt.title('Top 10 Products by Revenue')
plt.tight_layout()
plt.show()
plt.savefig('top10_products.png', bbox_inches='tight')
print("Saved top10_products.png")
```

```python
region_col = 'Region'

if region_col in df.columns:
    reg = df.groupby(region_col)[sales_col].sum().reset_index().sort_values(sales_col, ascending=False)
    plt.figure(figsize=(7,4))
    sns.barplot(data=reg, x=region_col, y=sales_col)
    plt.title('Sales by Region')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    plt.savefig('sales_by_region.png', bbox_inches='tight')
    print("Saved sales_by_region.png")
else:
    print(f"No column named {region_col}  skip region plot.")
```
